```python
import os
import geopy
import pandas
from geopy.geocoders import Nominatim
df7 = pandas.read_json("http://pythonhow.com/supermarkets.json")
df7=df7.set_index("ID")
df7["Continent"]=df7.shape[0]*["North America"]
df7_t=df7.T
df7_t["7"]=["My Address","Seattle", "USA", 1, "Newt's Emporium", "WA 98122", "North America"]
df7_t=df7_t.T
nom =Nominatim()
df7_t["Geocode"]=df7_t["Address"]+", "+df7_t["City"]+", "+df7_t["State"]+", "+df7_t["Country"]
df7_t["Coordinates"]=df7_t["Geocode"].apply(nom.geocode)
```

```python
df7_t["Lat"]=df7_t["Coordinates"].apply(lambda x: x.latitude if x !=None else None)
df7_t["Long"]=df7_t["Coordinates"].apply(lambda x: x.longitude if x !=None else None)
df7_t
```
